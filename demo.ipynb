```python
import parse
from bmc import BoundedModelChecker
from z3 import *
set_param(proof=True)
```

```python
with open('examples/short.smv') as f:
    input = f.read()
# Load and parse the NuSMV-like input file
smv_model = parse.parse(input)

# Create a bounded model checker
bmc = BoundedModelChecker(smv_model)

# Check the model
specification_not_satisfied = False

for k, l in [(k, l) for k in range(1, 5) for l in range(1, k+1)]:
    bmc.check(k, l)
    for modl in bmc.model:
        if modl is not None:
            print(f'Specification({k}, {l}) is not satisfied')
            bmc.print_model(k,modl)
            specification_not_satisfied = True
            break
    if specification_not_satisfied:
        break
else:
    print("Specification holds")
```
